# Neural Nets Notebook

Research on neural nets with simulated data.

```python
# libraries
import numpy as np
import sklearn
import sklearn.datasets
import sklearn.linear_model
import matplotlib
import matplotlib.pyplot as plt

# Display plots in notebook
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
# Simulation dataset for classification
N = 300
noise = 0.2

# can't give 'moons' zero noise
if(noise == 0): noise = None
    
x, y = sklearn.datasets.make_moons(n_samples=N, 
                                   shuffle=True, 
                                   noise = noise, 
                                   random_state=None)
```

```python
plt.scatter(x[:,0], # x is first column of x
            x[:,1], # y is second column of x
            s = 40, 
            c = y, # color based on response y vector
            cmap = plt.cm.Spectral)
```
